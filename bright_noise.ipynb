```python
from time_series import Collection, find_ders, find_files
import matplotlib.pyplot as plt
import numpy as np
```

```python
fname = "/home/lhillber/scratch/noise_trial_0.tdms" # Mount point
```

```python
X = Collection(fname, coord="x")
```

```python
D = X.collection[0]
X.apply("restore")
plt.loglog(*D.PSD())
plt.title(f"Raw data (r={D.r/1e6} MHz, N={D.Navg_psd} averages)");
```

```python
X.apply("restore")
plt.loglog(*X.average("PSD"))
plt.title(f"Spectral averaged data (r={X.r/1e6} MHz, N={X.Navg_psd} averages)");
```

```python
X.apply("restore")
plt.loglog(*X.average("PSD", taumax=10e-3))
plt.title(f"Block-spectral averaged data (r={round(X.r/1e6,1)} MHz, N={X.Navg_psd} averages)");
```

```python
X.apply("restore")
X.apply("bin_average", Npts=4, inplace=True)
plt.loglog(*X.average("PSD", taumax=10e-3))
plt.title(f"Bin-block-spectral averaged data (r={round(X.r/1e3,1)} kHz, N={X.Navg_psd} averages)");
```
